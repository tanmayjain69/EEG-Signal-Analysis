```
import mne
raw0 = mne.io.read_raw_edf("./drive/My Drive/dataset/B0101T.gdf", preload=True)
raw1 = mne.io.read_raw_edf("./drive/My Drive/dataset/B0102T.gdf", preload=True)
raw2 = mne.io.read_raw_edf("./drive/My Drive/dataset/B0103T.gdf", preload=True)
raw3 = mne.io.read_raw_edf("./drive/My Drive/dataset/B0104E.gdf", preload=True)
raw4 = mne.io.read_raw_edf("./drive/My Drive/dataset/B0105E.gdf", preload=True)
```

```
raw0.rename_channels({"EOG:ch01":"EOG1","EOG:ch02":"EOG2","EOG:ch03":"EOG3"})
raw0.rename_channels({"EEG:C3":"C3","EEG:Cz":"CZ","EEG:C4":"C4"})
raw0.info["ch_names"]
raw1.rename_channels({"EOG:ch01":"EOG1","EOG:ch02":"EOG2","EOG:ch03":"EOG3"})
raw1.rename_channels({"EEG:C3":"C3","EEG:Cz":"CZ","EEG:C4":"C4"})
raw1.info["ch_names"]
raw2.rename_channels({"EOG:ch01":"EOG1","EOG:ch02":"EOG2","EOG:ch03":"EOG3"})
raw2.rename_channels({"EEG:C3":"C3","EEG:Cz":"CZ","EEG:C4":"C4"})
raw2.info["ch_names"]
raw3.rename_channels({"EOG:ch01":"EOG1","EOG:ch02":"EOG2","EOG:ch03":"EOG3"})
raw3.rename_channels({"EEG:C3":"C3","EEG:Cz":"CZ","EEG:C4":"C4"})
raw3.info["ch_names"]
raw4.rename_channels({"EOG:ch01":"EOG1","EOG:ch02":"EOG2","EOG:ch03":"EOG3"})
raw4.rename_channels({"EEG:C3":"C3","EEG:Cz":"CZ","EEG:C4":"C4"})
raw4.info["ch_names"]
```

```
montage = mne.channels.read_montage("standard_1020")
raw0.set_montage(montage)
raw0.set_channel_types({'EOG1': 'eog'})
raw0.set_channel_types({'EOG2': 'eog'})
raw0.set_channel_types({'EOG3': 'eog'})

raw1.set_montage(montage)
raw1.set_channel_types({'EOG1': 'eog'})
raw1.set_channel_types({'EOG2': 'eog'})
raw1.set_channel_types({'EOG3': 'eog'})

raw2.set_montage(montage)
raw2.set_channel_types({'EOG1': 'eog'})
raw2.set_channel_types({'EOG2': 'eog'})
raw2.set_channel_types({'EOG3': 'eog'})

raw3.set_montage(montage)
raw3.set_channel_types({'EOG1': 'eog'})
raw3.set_channel_types({'EOG2': 'eog'})
raw3.set_channel_types({'EOG3': 'eog'})

raw4.set_montage(montage)
raw4.set_channel_types({'EOG1': 'eog'})
raw4.set_channel_types({'EOG2': 'eog'})
raw4.set_channel_types({'EOG3': 'eog'})
```

```
ica = mne.preprocessing.ICA(n_components=3, random_state=97, max_iter=800)
ica.fit(raw0)
ica.plot_properties(raw0)

ica.fit(raw1)
ica.plot_properties(raw1)

ica.fit(raw2)
ica.plot_properties(raw2)

ica.fit(raw3)
ica.plot_properties(raw3)

ica.fit(raw4)
ica.plot_properties(raw4)
```

```
raw0.plot_psd(fmax=250)
raw0.plot(duration=5, n_channels=3)

```

```
chs=['C3','CZ','C4']
chan_idxs = [raw0.ch_names.index(ch) for ch in chs]
raw0.plot(order=chan_idxs, start=12, duration=4)
```
